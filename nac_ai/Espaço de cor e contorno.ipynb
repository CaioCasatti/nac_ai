```python
#R1

%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('circulo.png')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)



image_lower_hsvBolaAzul = np.array([85, 160, 210])  
image_upper_hsvBolaAzul = np.array([95, 185, 230])

mask_hsvBolaAzul = cv2.inRange(img_hsv, image_lower_hsvBolaAzul, image_upper_hsvBolaAzul)

image_lower_hsvBolaVermelha1 = np.array([0, 110, 110])  
image_upper_hsvBolaVermelha1 = np.array([10, 240, 240])

mask_hsvBolaVermelha1 = cv2.inRange(img_hsv, image_lower_hsvBolaVermelha1, image_upper_hsvBolaVermelha1)

image_lower_hsvBolaVermelha2 = np.array([170, 110, 110])  
image_upper_hsvBolaVermelha2 = np.array([180, 240, 240])

mask_hsvBolaVermelha2 = cv2.inRange(img_hsv, image_lower_hsvBolaVermelha2, image_upper_hsvBolaVermelha2)

mask_hsvBolaVermelha = cv2.bitwise_or(mask_hsvBolaVermelha1, mask_hsvBolaVermelha2)


mask_hsvTotal = cv2.bitwise_or(mask_hsvBolaVermelha, mask_hsvBolaAzul)


contornos, _ = cv2.findContours(mask_hsvTotal, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) 
print("Contornos encontrados: ", len(contornos))

mask_rgbBolaAzul = cv2.cvtColor(mask_hsvBolaAzul, cv2.COLOR_GRAY2RGB) 
contornos_imgBolaAzul = mask_rgbBolaAzul.copy() 

cv2.drawContours(contornos_imgBolaAzul, contornos, -1, [100,50,0], 5);


mask_rgbBolaVermelha = cv2.cvtColor(mask_hsvBolaVermelha, cv2.COLOR_GRAY2RGB) 
contornos_imgBolaVermelha = mask_rgbBolaVermelha.copy() 

cv2.drawContours(contornos_imgBolaVermelha, contornos, -1, [100,50,0], 5);

contornos_img = cv2.bitwise_or(contornos_imgBolaAzul, contornos_imgBolaVermelha)

plt.figure(figsize=(8,6))
plt.imshow(contornos_img);

cnt = contornos[0]

M = cv2.moments(cnt)


cnt = contornos[1]
M2 = cv2.moments(cnt)


cxBolaVermelha = int(M['m10']/M['m00'])
cyBolaVermelha = int(M['m01']/M['m00'])


cxBolaAzul = int(M2['m10']/M2['m00'])
cyBolaAzul = int(M2['m01']/M2['m00'])

print("centro de massa na posição (Bola Azul): ",cxBolaAzul, cyBolaAzul)
print("centro de massa na posição (Bola Vermelha): ",cxBolaVermelha, cyBolaVermelha)

size = 20
color = (31, 85, 72)

cv2.line(contornos_imgBolaVermelha,(cxBolaVermelha - size,cyBolaVermelha),(cxBolaVermelha + size,cyBolaVermelha),color,5)
cv2.line(contornos_imgBolaVermelha,(cxBolaVermelha,cyBolaVermelha - size),(cxBolaVermelha, cyBolaVermelha + size),color,5)


cv2.line(contornos_imgBolaAzul,(cxBolaAzul - size,cyBolaAzul),(cxBolaAzul + size,cyBolaAzul),color,5)
cv2.line(contornos_imgBolaAzul,(cxBolaAzul,cyBolaAzul - size),(cxBolaAzul, cyBolaAzul + size),color,5)

cv2.line(contornos_img,(cxBolaAzul,cyBolaAzul),(cxBolaVermelha,cyBolaVermelha),color,5)


font = cv2.FONT_HERSHEY_SIMPLEX
textBolaAzul = cxBolaAzul , cyBolaAzul
origemBolaAzul = (700,50)

font = cv2.FONT_HERSHEY_SIMPLEX
textBolaVermelha = cxBolaVermelha , cyBolaVermelha
origemBolaVermelha = (700,150)



cv2.putText(contornos_imgBolaAzul, str(textBolaAzul), origemBolaAzul, font,1,(73, 203, 230),2,cv2.LINE_AA)

cv2.putText(contornos_imgBolaVermelha, str(textBolaVermelha), origemBolaVermelha, font,1,(100,50,0),2,cv2.LINE_AA)

contornos_img = cv2.bitwise_or(contornos_imgBolaAzul, contornos_imgBolaVermelha)


plt.imshow(contornos_img);



```
